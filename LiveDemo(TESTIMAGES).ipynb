```python
# IMPORTS
import os
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import cv2

from google.colab import drive
drive.mount('/content/drive')


from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.models import load_model
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' # Supress tensorflow log outputs
```

```python
### TEST CASE: AGARWOOD ###

### (VERY IMPORTANT) INSERT FILE PATHS ###
model_path = ("/content/drive/MyDrive/LIVE DEMO/bcshs-agarwood-MobileNetV2.keras")
image_path = ("/content/drive/MyDrive/LIVE DEMO/TEST FILES/test(agar).jpg")



# VARIABLES
model = load_model(model_path)
img_size = (299, 299)
batch_size = 32

#LOADS IMAGE
image = load_img(image_path, target_size=img_size)
input_arr = img_to_array(image)
input_batch = np.expand_dims(input_arr, axis=0)
pred = model.predict(input_batch)

CLASS_NAMES = ["Agarwood", "Non-Agarwood"]

#Test Image Visualization
# Reading an image in default mode
img = cv2.imread(image_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #Converting BGR to RGB
# Displaying the image
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(299,299))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Convert single image to a batch.
predictions = model.predict(input_arr)
result_index = np.argmax(predictions) #Return index of max element
print(result_index)
predicted_class = CLASS_NAMES[result_index]
print("It's an {}".format(predicted_class))
print(predictions)
```
